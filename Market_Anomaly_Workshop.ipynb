## Correlation Analysis

2. Run correlation
3. Plot Heatmap to easily recognize correlations within the data

```python
# get columns with numeric column data types
numeric_df = df.select_dtypes(include=[np.number])
# run a correlation across the numeric dataframe to see how data relates to
# each other
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
# Get the upper triangle of the correlation matrix to avoid duplicates
upper_tri = np.triu(corr_matrix)

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix), k = 1)

# Set up the matplotlib figure
plt.figure(figsize=(12,8))

# Create heatmap using seaborn
sns.heatmap(corr_matrix,
            mask=mask,
            annot=False,
            cmap='coolwarm',
            vmin=-1,
            vmax=1,
            center=0)

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Import Isolation Forest
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report

# Prepare features (X)
X = df[['VIX']]
y = df['Y']

# Create and train the isolation forest model
iso_forest = IsolationForest(
    n_estimators=200, # Number of trees (what is fit to the dataset)
    contamination=0.2, # Expected proportion of outliers
    random_state=42,
    n_jobs=-1 # Use all CPU cores
)

# Fit and predict
# predict() returns 1 for inliers and -1 for outliers
predictions = iso_forest.fit_predict(X)

# Convert predictions to binary format (0 for normal and 1 for anomaly)
# Keeps labels consistent
y_pred = np.where(predictions == -1, 1, 0)

# Print classification report
print("Classification Report")
print(classification_report(y,y_pred))

# Calculate and print anomaly scores
anomaly_scores = iso_forest.score_samples(X)
print("\nAnomaly Score Stats:")
print(f"Mean Score: {np.mean(anomaly_scores):.4f}")
print(f"Min Score: {np.min(anomaly_scores):.4f}")
print(f"Max Score: {np.max(anomaly_scores):.4f}")
```

## Using Time/Dates in your data (moving avgs)

```python
def create_time_split(df, train_years=14, validation_weeks=52, test_weeks=52):
  """
  Specialized split for financial time series

  Args:
    df: DataFrame with weekly data
    train_years: Number of years for training (default 3 years)
    validation_weeks: Weeks for validation (default 26 weeks = 6 months)
    test_weeks: Weeks for testing
  """

  # Convert date columne to datetime
  df['Data'] = pd.to_datetime(df['Data'])
  df = df.sort_values('Data')

  # Calculate sizes
  train_size = train_years * 52
  total_required_size = train_size + validation_weeks + test_weeks

  if len(df) < total_required_size:
    raise ValueError("Not enough data for specified split size")

  # Create Splits
  train_end = train_size
  val_end = train_end + validation_weeks

  train_data = df.iloc[:train_end]
  validation_data = df.iloc[train_end:val_end]
  test_data = df.iloc[val_end:val_end + test_weeks]

  # Verify Splits
  print(f"Training Period: {train_data['Data'].min()} to {train_data['Data'].max()}")
  print(f"Validation Period: {validation_data['Data'].min()} to {validation_data['Data'].max()}")
  print(f"Testing Period: {test_data['Data'].min()} to {test_data['Data'].max()}")

  return train_data, validation_data, test_data
```

```python
# Split the data
train_data, val_data, test_data = create_time_split(df)
```

```python
# Scale the features using StandardScaler
from sklearn.preprocessing import StandardScaler
# Prepare features (X) and target (y)
X_train = train_data[['VIX', 'GTITL2YR', 'GTDEM2Y', "EONIA", 'GTJPY2YR', 'DXY']]
X_test = test_data[['VIX', 'GTITL2YR', 'GTDEM2Y', "EONIA", 'GTJPY2YR', 'DXY']]
y_train = train_data['Y']
y_test = test_data['Y']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the model
model = LogisticRegression(
    random_state=42,
    max_iter=1000, # Increase iterations to ensure convergence
    C=0.15, # Stronger regularization to prevent overfitting
    class_weight='balanced', # Handle class imbalance (how non-anomalies have more data than anomalies)
    solver='lbfgs', # Efficient solver for small datasets
    n_jobs=-1 # Use all CPU cores for faster creation and training
    )

model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Print classification report
print('Classification Report')
print(classification_report(y_test, y_pred))

#Print model coefficients
print("\nModel Coefficients (Scaled Features):")
print(f"VIX coefficients: {model.coef_[0][0]:.4f}")
print(f"Intercept: {model.intercept_[0]:.4f}")
```

## Train model with moving average

```python
# To include moving average
df["VIX_MA5"] = train_data['VIX'].rolling(window=5).mean() #this is a 5 year moving average
df["VIX_MA10"] = df['VIX'].rolling(window=10).mean()
df["VIX_MA20"] = df['VIX'].rolling(window=20).mean()
df["VIX_MA_CROSS"] = (df["VIX_MA5"] > df["VIX_MA20"]).astype(int)

X = df[['VIX', 'VIX_MA_CROSS']]
y = df['Y']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.4, random_state=42)
# Create and train the model
model = LogisticRegression(
    random_state=42,
    max_iter=1000, # Increase iterations to ensure convergence
    C=0.15, # Stronger regularization to prevent overfitting
    class_weight='balanced', # Handle class imbalance (how non-anomalies have more data than anomalies)
    solver='lbfgs', # Efficient solver for small datasets
    n_jobs=-1 # Use all CPU cores for faster creation and training
    )

model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Print classification report
print('Classification Report')
print(classification_report(y_test, y_pred))

#Print model coefficients
print("\nModel Coefficients (Scaled Features):")
print(f"VIX coefficients: {model.coef_[0][0]:.4f}")
print(f"Intercept: {model.intercept_[0]:.4f}")
```
